```python
df_aex['localSymbolMonth'] = df_aex['multiplier'][:2]
df_aex.head()
```

```python
unique_set = set(pd.concat([aex_schedule.apply(lambda aex_schedule: aex_schedule['CARRY_CONTRACT'][4:], axis=1),
                            aex_schedule.apply(lambda aex_schedule: aex_schedule['PRICE_CONTRACT'][4:], axis=1)]))
    
unique_set
```

```python
df_contracts = df_aex[df_aex['localSymbolMonth'].isin(get_month_symbols(unique_set))]
```

```python
pd.to_datetime(df_contracts.iloc[0]['lastTradeDateOrContractMonth'][:8])
```

```python
df_contracts
```

```python
import datetime
datetime.datetime.today() - datetime.timedelta(2)
```

```python
for row in df_contracts.itertuples():
    contract = Contract(conId=row.conId, includeExpired=True)
    ib.qualifyContracts(contract)
    last_trade_date = pd.to_datetime(df_contracts.loc[row.Index]['lastTradeDateOrContractMonth'][:8])
    last_trade_date_str = last_trade_date.strftime("%Y%m%d %H:%M:%S")
    head_date = ib.reqHeadTimeStamp(contract, whatToShow='TRADES', useRTH=True)
    head_date = head_date + datetime.timedelta(365)
    head_date_str = head_date.strftime("%Y%m%d %H:%M:%S")
    end_date = row.lastTradeDateOrContractMonth
    end_date = end_date + " 00:00:00"
    print('head_date_str: ', head_date_str, "last_trade_date_str: ", last_trade_date_str)  
    bars = ib.reqHistoricalData(
        contract,
        last_trade_date_str,
        #'20180318 00:00:00',
        #end_date,
        durationStr='365 D',
        barSizeSetting='1 day',
        whatToShow='TRADES',
        useRTH=True,
        formatDate=1)
    df = util.df(bars)
    df.tail()
   
```
